```python
import ROOT
from math import radians,cos,sin,tan, pi
import numpy as np
import os
import pandas as pd
from array import array
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from skimage.io import imread, imshow
from skimage import data_dir
from skimage.transform import radon, iradon, iradon_sart
```

```python
path_data = "../Data/Rates/"
```

```python
alpha_giro = 160
radius = 32

#thetas =np.array([187.5,210,165])
thetas =np.array([157.5,142.5,187.5,210,165])
thetas_rad = [radians(i) for i in thetas]

alphas = np.arange(0,360,360/alpha_giro)
alphas_rad = [radians(i) for i in alphas]
```

```python
rates = []
for i in range(1,6):
    s = str(i)
    name_file = "Positioning"+s
    r = np.load(path_data + name_file +"/rate_rel_pos"+s+".npy")
    rates.append(r)

rates = np.array(rates)
```

```python
data_sinogram = [[],[],[]]

for i in range(0,len(thetas_rad)):
    for j in range(0,len(alphas_rad)):
        phi = (thetas_rad[i]/2)-alphas_rad[j]
        if(i==3):
            continue
        else:
            if phi>=-pi/2 and phi<pi/2:
                data_sinogram[0].append(radius*abs(cos(thetas_rad[i]/2)))
                data_sinogram[1].append(phi)
                data_sinogram[2].append(rates[i][j])
                #print(i,thetas[i])
            else:
                data_sinogram[0].append(-radius*abs(cos(thetas_rad[i]/2)))
                data_sinogram[1].append(phi)
                data_sinogram[2].append(rates[i][j])
            #print("i",i,theta[i])

```

```python
d_sinogram = np.array(data_sinogram)
d_sinogram = d_sinogram.transpose()
df_sinogram = pd.DataFrame(d_sinogram,columns = ["s","phi","rates"])
df_sinogram.head()
```

```python
nbinx = 60
nbiny = 120
inf_x = -11
sup_x = 11
inf_y = -2*pi
sup_y = pi
dimx = (sup_x-inf_x)/nbinx
dimy = (sup_y-inf_y)/nbiny
passo = (sup_x - inf_x) / nbinx

c1 = ROOT.TCanvas("c1","c1",1000,1000,1000,1000)
h2_reconstruction = ROOT.TH2F("h2","h2",nbinx,inf_x,sup_x,nbiny,inf_y,sup_y)
```

```python
for i in range(0,alpha_giro*len(thetas_rad)):
        h2_reconstruction.Fill(df_sinogram["s"][i],df_sinogram["phi"][i],df_sinogram["rates"][i])
```

```python
ROOT.gStyle.SetPalette(ROOT.kInvertedDarkBodyRadiator)
h2_reconstruction.Draw("COLZ")
h2_reconstruction.SetStats(0)

c1.Print("Reconstruction.png","png")
```
