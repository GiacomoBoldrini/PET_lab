## Senza Dense measures

```python
path_data = "/Users/boldrinicoder/lab4/Working_sinogram/Rates/"
path_graph = "/Users/boldrinicoder/lab4/Working_sinogram/Working_sin_graphs/"
```

```python
alpha_giro = 160
radius = 32
alpha_iniziale_sorgente = 4.5*pi/180

#Inserisco i theta e li trasformo tutti in angoli <pi

#Parte da tenere quando non hai tutti i positioning
#thetas_corr = np.array([157.5,142.5,187.5,210,165])
#thetas = np.array([157.5,142.5,165])

thetas = np.array([157.5,142.5,172.5,150,165])



thetas_rad = [radians(i) for i in thetas]

#Inserisco gli alpha
alphas = np.arange(0,360,360/alpha_giro)
alphas_rad = [radians(i) for i in alphas]
```

```python
rates = []
for i in range(1,6):
    #if i==4:
       # continue
    s = str(i)
    name_file = "Positioning"+s
    r = np.load(path_data + name_file +"/rate_rel_pos"+s+".npy")
    rates.append(r)

rates = np.array(rates)
```

```python
data_sinogram = [[],[],[]]

for i in range(0, len(thetas_rad)):
    for j in range(0, len(alphas_rad)):
        phi = thetas_rad[i]/2 - alphas_rad[j]

        data_sinogram[0].append(radius*cos(thetas_rad[i]/2))
        data_sinogram[1].append(phi)
        data_sinogram[2].append(rates[i][j])
```

```python
d_sinogram = np.array(data_sinogram)
d_sinogram = d_sinogram.transpose()
df_sinogram = pd.DataFrame(d_sinogram,columns = ["s","phi","rates"])
df_sinogram.head()
```

```python
#Creo il sinogramma
##Nel frattempo creo un dataframe con tutti i dati lavorati utile
#per il TGraph dopo
final_data=[[],[],[]]

nbinx = 100
nbiny = 100
inf_x = -11
sup_x = 11
inf_y = 0
sup_y = pi
dimx = (sup_x-inf_x)/nbinx
dimy = (sup_y-inf_y)/nbiny
passo = (sup_x - inf_x) / nbinx

c = ROOT.TCanvas("c","c",1000,1000,1000,1000)
h_reconstruction = ROOT.TH2F("h","h",nbinx,inf_x,sup_x,nbiny,inf_y,sup_y)

for i in range(0,alpha_giro*len(thetas_rad)):
    if df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente < -pi:
        if 2*pi + df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente > 0:
            h_reconstruction.Fill(df_sinogram["s"][i],2*pi + df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente,df_sinogram["rates"][i])
            final_data[0].append(df_sinogram["s"][i])
            final_data[1].append(2*pi + df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente)
            final_data[2].append(df_sinogram["rates"][i])
        else:
            h_reconstruction.Fill(-df_sinogram["s"][i],2*pi + df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente + pi,df_sinogram["rates"][i])
            final_data[0].append(-df_sinogram["s"][i])
            final_data[1].append(2*pi + df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente + pi)
            final_data[2].append(df_sinogram["rates"][i])
    else:
        if df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente > 0:
            h_reconstruction.Fill(df_sinogram["s"][i],df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente,df_sinogram["rates"][i])
            final_data[0].append(df_sinogram["s"][i])
            final_data[1].append(df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente)
            final_data[2].append(df_sinogram["rates"][i])
        else:
            h_reconstruction.Fill(-df_sinogram["s"][i],df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente +pi,df_sinogram["rates"][i])
            final_data[0].append(-df_sinogram["s"][i])
            final_data[1].append(df_sinogram["phi"][i] + pi/2 - alpha_iniziale_sorgente + pi)
            final_data[2].append(df_sinogram["rates"][i])
```

```python
final_data_array = np.array(final_data)
final_data_array = final_data_array.transpose()
df_final_data = pd.DataFrame(final_data_array, columns = ["final_s","final_phi","rates"])
```

```python
ROOT.gStyle.SetPalette(ROOT.kInvertedDarkBodyRadiator)
h_reconstruction.Draw("COLZ")
h_reconstruction.SetStats(0)

c.Draw("")
```

```python
#Plotto il TGraph
c_graph = ROOT.TCanvas("c_graph","c_graph",1000,1000,1000,1000)
graph = ROOT.TGraph2D()
graph.SetNpy(100)
graph.SetNpx(100)

for i in range(0, len(df_final_data["final_s"])):
    graph.SetPoint(i, df_final_data["final_s"][i], df_final_data["final_phi"][i], df_final_data["rates"][i])
    
    

graph.GetXaxis().SetTickLength(0.)
graph.GetYaxis().SetTickLength(0.)    
graph.Draw("COLZ")
c_graph.Draw("")
c_graph.Print("Graph_da_ritagliare.png", ".png")
c_graph.Print("Graph_da_ritagliare.root", "root")
```
